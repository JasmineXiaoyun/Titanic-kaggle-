```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
plt.rcParams['font.sans-serif']=['SimHei']#正常显示中文
```

```python
train_file=pd.read_csv("data/train.csv")
train_file
```

```python
train_file.info()#可以看出feature的缺失比例
```

```python
train_file.describe()
```

```python
#获救与仓位的关系
survived_0=train_file.Pclass[train_file.Survived==0].value_counts()
survived_1=train_file.Pclass[train_file.Survived==1].value_counts()
df=pd.DataFrame({'获救':survived_1,"未获救":survived_0})
df.plot(kind='bar',stacked=True)
```

```python
#获救与性别的关系
survived_0=train_file.Sex[train_file.Survived==0].value_counts()
survived_1=train_file.Sex[train_file.Survived==1].value_counts()
df=pd.DataFrame({'获救':survived_1,"未获救":survived_0})
df.plot(kind='bar',stacked=True)
```

```python
#获救与年龄的关系
survived_0=train_file.Age[train_file.Survived==0].value_counts()
survived_1=train_file.Age[train_file.Survived==1].value_counts()
df=pd.DataFrame({'获救':survived_1,"未获救":survived_0})
df.plot(kind='line')
```

```python
#获救与sibsp的关系
survived_0=train_file.SibSp[train_file.Survived==0].value_counts()
survived_1=train_file.SibSp[train_file.Survived==1].value_counts()
df=pd.DataFrame({'获救':survived_1,"未获救":survived_0})
df.plot(kind='bar',stacked=True)
```

```python
#获救与parch的关系
survived_0=train_file.Parch[train_file.Survived==0].value_counts()
survived_1=train_file.Parch[train_file.Survived==1].value_counts()
df=pd.DataFrame({'获救':survived_1,"未获救":survived_0})
df.plot(kind='bar',stacked=True)
```

```python
#获救与Fare的关系
survived_0=train_file.Fare[train_file.Survived==0].value_counts()
survived_1=train_file.Fare[train_file.Survived==1].value_counts()
df=pd.DataFrame({'获救':survived_1,"未获救":survived_0})
plt.scatter(train_file.Survived,train_file.Fare)
```

```python
#获救与Cabin的关系
survived_cabin=train_file.Survived[pd.notnull(train_file.Cabin)].value_counts()
survived_nocabin=train_file.Survived[pd.isnull(train_file.Cabin)].value_counts()
df=pd.DataFrame({'有舱':survived_cabin,"无舱":survived_nocabin}).transpose()
df.plot(kind='bar',stacked=True)
```

```python
#获救与Embarked的关系
survived_0=train_file.Embarked[train_file.Survived==0].value_counts()
survived_1=train_file.Embarked[train_file.Survived==1].value_counts()
df=pd.DataFrame({'获救':survived_1,"未获救":survived_0})
df.plot(kind='bar',stacked=True)
```

```python
#把文字转换为数字
sex_mapping={"male":0,"female":1}
train_file.Sex=train_file.Sex.map(sex_mapping)

embarked_mapping={"S":0,"C":1,"Q":2}
train_file.Embarked=train_file.Embarked.map(embarked_mapping)

train_file.loc[(train_file.Cabin.notnull()),'Cabin']=1
train_file.loc[(train_file.Cabin.isnull()),'Cabin']=0
 
#去掉不用的两列
train_data=train_file.drop(columns=['Name','Ticket'])

#把feature缺失值填为-1
train_data=train_data.fillna(-1)
```

```python
train_data
```

```python
#拆分训练集测试集
train_data_array=np.array(train_data)
print(train_data_array)
print(" ")
np.random.shuffle(train_data_array)#随机打乱
print(train_data_array)
print(" ")
train_set=train_data_array[:int(0.8*len(train_data))]
test_set=train_data_array[int(0.8*len(train_data)):]
print(train_set[:100])
```

```python
#将array转为dataframe
df_train=pd.DataFrame(train_set,columns=['PassengerId','Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked'])
df_test=pd.DataFrame(test_set,columns=['PassengerId','Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked'])
print(df_train)
print(df_test)

df_train_y=df_train.loc[:,'Survived']
print(df_train_y)
df_train_x=df_train.drop(['Survived'],axis=1)
print(df_train_x)
df_test_y=df_test.loc[:,'Survived']
print(df_test_y)
df_test_x=df_test.drop(['Survived'],axis=1)
print(df_test_x)
```

```python
#建模
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model=LogisticRegression()
model.fit(df_train_x,df_train_y)
```

```python
#训练集
predicted_classes_train=model.predict(df_train_x)
accuracy_train=accuracy_score(df_train_y.values,predicted_classes_train)
print(accuracy_train)
predicted_classes_test=model.predict(df_test_x)
accuracy_test=accuracy_score(df_test_y.values,predicted_classes_test)
print(accuracy_test)
```

```python
#测试集
#data processing
test_file=pd.read_csv("data/test.csv")

#把文字转换为数字
sex_mapping={"male":0,"female":1}
test_file.Sex=test_file.Sex.map(sex_mapping)

embarked_mapping={"S":0,"C":1,"Q":2}
test_file.Embarked=test_file.Embarked.map(embarked_mapping)

test_file.loc[(test_file.Cabin.notnull()),'Cabin']=1
test_file.loc[(test_file.Cabin.isnull()),'Cabin']=0
 
#去掉不用的两列
test_data=test_file.drop(columns=['Name','Ticket'])

#把feature缺失值填为-1
test_data=test_data.fillna(-1)
```

```python
predicted_classes=model.predict(test_data)
predicted_classes
```
